```python
import numpy as np
from scipy.optimize import root_scalar


def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

def df(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27

interval = [-100, 100]
step = 0.5

def find_real_roots_scipy(interval, step, tol=1e-10):
    roots = []
    x_values = np.arange(interval[0], interval[1], step)
    
    for x in x_values:
        try:
            sol = root_scalar(f, x0=x, fprime=df, method='newton')
            root = sol.root
            if sol.converged and np.isreal(root):
                root = np.round(root, decimals=10).real
                if not any(np.isclose(root, r, atol=tol) for r in roots):
                    roots.append(root)
                   
        except (RuntimeError, OverflowError):
            pass
    
    roots.sort()
    return roots

real_roots = find_real_roots_scipy(interval, step)
print("Raíces reales en orden ascendente:", real_roots)
```
